### Going inside the website and extracting all the markup language from that and then trying to get the Clinic I'D with Clinic Name from the particular page

```python
import requests
from bs4 import BeautifulSoup
```

##### Extracting information from the website: "https://19695.portal.athenahealth.com/"
- And seaching CLinic Name from their given I'D which is given in URL:
- For exmaple Clinic I'D = 19695 and Clinic Name : Atlas OB-GYN

```python
def get_clinic_name(clinic_id):
    start_url = f'https://{clinic_id}.portal.athenahealth.com/'
    response = requests.get(start_url)
    soup = BeautifulSoup(response.text,'html.parser')
    clinic_name = soup.find_all('h1')[-1].text.strip()
    return clinic_name
```

```python
start_from = 1 # starting from 
end_at = 500   #  ending at, you can increase the number 
master_list=[]
for clinic_id in range(start_from,end_at+1):
    data_dict = {}
    data_dict['clinic_id'] = clinic_id
    data_dict['clinic_name'] = get_clinic_name(clinic_id)
    if data_dict['clinic_name']!= 'Payment Confirmation' and data_dict['clinic_name']!= "Sorry, we can't find that practice. Make sure you typed the right address.":
        master_list.append(data_dict)
```

#### Converting all the data which we stored in list named as master_list into CSV file

```python
import pandas as pd
df = pd.DataFrame(master_list)
df.to_csv('clinic_data.csv',index=False)
```
