# Downloading All the images from the Webpage : https://www.pngmart.com

#### Total images in the website is : 11,921 *20 = 238420

##### 238420 Images in the websites and all the images are downloaded using this code...

```python
# Importing all important libraries which are required to download images
import requests
from bs4 import BeautifulSoup
```

```python
site_map  = 'https://www.pngmart.com/sitemap.xml'  # taking website's sitemap
response  = requests.get(site_map)
soup      = BeautifulSoup(response.text,'xml')
soup
```

```python
# Extracted all the important pages links
site_map = []
for page_links in soup.find_all('loc'):
    posts_url = page_links.text
    if 'posts' in posts_url:
        site_map.append(posts_url)
site_map
```

```python
# Downloading all the images from the sites
for links in site_map:
    print(links)
    response =requests.get(links)
    soup = BeautifulSoup(response.text,'xml')
    for image_links in soup.find_all('loc'):
        image_links =  image_links.text
        if 'image' in image_links:
            response =requests.get(image_links)
            soup = BeautifulSoup(response.text,'xml')
            png_url = soup.find('a',{'class':'download'})['href']
            image = requests.get(png_url)
            png_title = image_links.split("/")[-1]+"-"+soup.find('a',{'class':'download'})['href'].split("/")[-1]
            print(png_title)
            with open (png_title,'wb') as file:
                file.write(image.content)
            print("print Next Image")
    print("Next pages")
            
# Interrupted it because it was taking to long time, as there are 2 lakhs above photos
```
